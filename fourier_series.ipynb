```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
omega = 2 * np.pi
t = np.linspace(0, 2, 2000)
N = 15  # highest odd harmonic

# Odd harmonics
f = np.zeros_like(t)
for n in range(1, N+1, 2):
    f += np.cos(n * omega * t)

# Scaling
f *= 16 / np.pi

# Plot
plt.figure(figsize=(20,4))
plt.plot(t, f)
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title(r"$\frac{16}{\pi}\sum_{n=1,3,5,\dots}^{%d} \cos(n\omega t)$" % N)
plt.grid(True)
plt.show()
```

```python


# Parameters
A = 1
T = 2
omega = 2 * np.pi / T
t = np.linspace(-T, T, 4000)

N = 3 # number of harmonics (odd only)

# Fourier series
x = np.zeros_like(t)

for n in range(1, N+1, 2):
    x += ((-1)**((n-1)//2)) * (1/n) * np.cos(n * omega * t)

x *= 4 * A / np.pi

# Plot
plt.figure(figsize=(20,4))
plt.plot(t, x)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Fourier Series Approximation of Given Square Wave")
plt.grid(True)
plt.show()
```
